## lookahead, lookbehind

```python
for item in re.finditer('(?P<title>[\w ]+)(?=\[edit\])', wiki):
    print(item)
```

```python
for item in re.finditer('(?P<title>[\w ]+)(?=\[edit\])', wiki):
    print(item.groupdict()['title'])
```

| Type | Name | Syntax | Meaning |
|-----:|------|--------|---------|
| ➕ | Positive Lookahead | `X(?=Y)` | Match X only if followed by Y |
| ➖ | Negative Lookahead | `X(?!Y)` | Match X only if NOT followed by Y |
| ⬅️➕ | Positive Lookbehind | `(?<=Y)X` | Match X only if preceded by Y |
| ⬅️➖ | Negative Lookbehind | `(?<!Y)X` | Match X only if NOT preceded by Y |


```python
with open ('datasets/buddhist.txt', 'r', encoding='utf-8') as file:
    wiki = file.read()

wiki
```

```python
pattern = """
(?P<title>.*)            #the title of the uni
(–\ located\ in\ )         # an indicator of loc
(?P<city>\w*)            # city of the uni
(,\ )                     # separator for state
(?P<state>\w*)           # state of the uni
"""

for item in re.finditer(pattern, wiki, re.VERBOSE): #verbose fontos!
    print(item.groupdict())
```
