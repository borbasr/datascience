# Mennyi kilometert tudnak megtenni engerszam alapjan

```python
CtyMpgByCyl = []

for c in cylinders:
    summpg=0
    cyltpercount=0
    for d in mpg:
        if d['cyl'] == c:
            summpg += float(d['cty'])
            cyltpercount+=1
    CtyMpgByCyl.append((c, summpg / cyltpercount))
CtyMpgByCyl.sort(key=lambda x: x[0])
CtyMpgByCyl
```

```python
vehicleclass = set(d['class'] for d in mpg)
vehicleclass
```

```python
HwyMpgByCyl = []

for c in vehicleclass:
    summpg=0
    cyltpercount=0
    for d in mpg:
        if d['class'] == c:
            summpg += float(d['cty'])
            cyltpercount+=1
    HwyMpgByCyl.append((c, summpg / cyltpercount))
HwyMpgByCyl.sort(key=lambda x: x[0])
HwyMpgByCyl
```

```python
import re
```

```python
text = 'Amy works diligently. Amy gets good graders. our student Amy usccesful.'
re.split('Amy', text)
```

```python
re.search('^Amy', text)
```

```python
print(re.match('^Amy', text)
```
