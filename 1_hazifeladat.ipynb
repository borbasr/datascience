## 1. feladat, Regex

A nytimeshealth.txt filet használva,

    1.1. Keresd ki az összes weblap címet (url)
    1.2. Keresd ki az összes twitter felhasználónevet
    1.3. Keresd ki az összes főcímet
    1.4. Add meg egy szótárba, hogy havi bontásban hány bejegyzés található.




## 2. feladat, Numpy Tömb

    2.1. Hozz létre egy (100, 5) alakú NumPy tömböt (array), amely 0 és 1000 közötti véletlenszerű egész számokat tartalmaz. Minden sor egy külön mintát, minden oszlop egy külön jellemzőt (feature) képvisel.
    2.2 Számítsd ki és írd ki az egyes jellemzők átlagát, mediánját és szórását.
    2.3 Normalizáld az adatokat úgy, hogy minden jellemző átlaga 0 és szórása 1 legyen.
    2.4 Keresd meg és írd ki azokat a mintákat, ahol az első jellemző nagyobb mint 500 és a második jellemző kisebb mint 200.
    2.5 Hozz létre egy új jellemzőt, amely az első és harmadik jellemzők összege, és add hozzá a tömbhöz új oszlopként.
    2.6 Alakítsd át a tömböt egy 20 soros és 30 oszlopos tömbbé.
    2.7 Mentsd el az átalakított tömböt egy fájlba, majd töltsd be újra egy új tömbbe.
    2.8 Keresd meg és írd ki az új tömb egyedi értékeit és azok előfordulási számát.
